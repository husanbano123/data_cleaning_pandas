# 2. What are the orders with missing ProductIDs?
we have to find missing values in ProductID or CustomerID,TransactionID by isna() method.if we take df[] then it will return the result in tabular dataframe format.and here there is no missing values are present in productID

```python
missing_order_ids = df[df['CustomerID'].isna()] #there is no missing id
missing_order_ids
```

#  8. Which product is most popular?
# The idxmax() method returns a Series with the index of the maximum value for each column. 

```python
most_popular_product=df.groupby('ProductID')['Quantity'].sum().idxmax()
most_popular_product
```

```python
most_popular_product=df.groupby('Country')['Quantity'].sum().idxmax()
most_popular_product
```
